```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Control Cell

```python
# controla is the results from a control cell
controla = pd.read_csv("/Users/BrittanyDorsey/Desktop/my-notebook/MyRepo/ornet-reu-2018/results/results_DsRed2-HeLa_10_26_Cell0.csv")
# controlb is the ms_results from the same control cell
controlb = pd.read_csv("/Users/BrittanyDorsey/Desktop/my-notebook/MyRepo/ornet-reu-2018/results/MS_results_DsRed2-HeLa_10_26_Cell0.csv")
```

```python
FSVa = controla['FSV']
FSVb = controlb['FSV']

pa = controla['pval']
pb = controlb['pval']

qa = controla['qval']
qb = controlb['qval']

la = controla['LLR']
lb = controlb['LLR']
```

```python
FSVa2 = FSVa.values.reshape(512, 512)
FSVb2 = FSVb.values.reshape(512, 512)

pa2 = pa.values.reshape(512, 512)
pb2 = pb.values.reshape(512, 512)

qa2 = qa.values.reshape(512, 512)
qb2 = qb.values.reshape(512, 512)

la2 = la.values.reshape(512, 512)
lb2 = lb.values.reshape(512, 512)
```

```python
FSVa2.shape
```

```python
# Results
plt.subplot(2, 2, 1)
plt.imshow(FSVa2)
plt.title("Fraction Spatial Variance")
plt.colorbar()
plt.subplot(2, 2, 2)
plt.imshow(pa2)
plt.title("P-value")
plt.colorbar()
plt.subplot(2, 2, 3)
plt.imshow(qa2)
plt.title("Q-value")
plt.colorbar()
plt.subplot(2, 2, 4)
plt.imshow(la2)
plt.title("Log-likelihood Ratio")
plt.colorbar()
```

```python
# MS Results
plt.subplot(2, 2, 1)
plt.imshow(FSVb2)
plt.colorbar()
plt.subplot(2, 2, 2)
plt.imshow(pb2)
plt.colorbar()
plt.subplot(2, 2, 3)
plt.imshow(qb2)
plt.colorbar()
plt.subplot(2, 2, 4)
plt.imshow(lb2)
plt.colorbar()
```

```python
plt.subplot(2, 2, 1)
plt.hist(FSVa)
plt.subplot(2, 2, 2)
plt.hist(pa)
plt.subplot(2, 2, 3)
plt.hist(qa)
plt.subplot(2, 2, 4)
plt.hist(la)
```
